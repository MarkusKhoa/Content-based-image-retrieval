```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_origin = mpimg.imread("sample_image.jpg")
plt.axis("off")
plt.imshow(img_origin)
```

## Problem

### Same image but different brightness

```python
a, b = 0.05, 0.02
img_transform = img_origin * a - b
```

```python
_, axes = plt.subplots(1, 2, figsize = (10, 10))
axes = axes.flatten()
for name, img, ax in zip(["Original image", "Transformed image"], [img_origin, img_transform], axes):
    ax.imshow(img)
    ax.axis("off")
    ax.set_title(name)

plt.show()
```

```python
mse = ((img_origin - img_transform) ** 2).sum()
print(mse)
```

### Solving problem by using cosine similarity

```python
a, b = 0.05, 0.02
img_transform = img_origin * a - b

_, axs = plt.subplots(1, 2, figsize=(8, 8))
axs = axs.flatten()
for name, img, ax in zip(["image_origin", "image_tranform"], [img_origin, img_transform], axs):
    ax.imshow(img)
    ax.set_title(name)
plt.show()
```

```python
def normalize(x, axis = None):
    return np.sqrt(np.sum(x ** 2, axis = axis, keepdims = True))
```

```python
norm_img_origin = img_origin / normalize(img_origin, axis = (0, 1))
norm_img_transform = img_transform / normalize(img_transform, axis = (0, 1))
```

```python
np.abs(norm_img_origin - norm_img_transform).sum()
```
